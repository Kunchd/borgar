```python
from scipy.stats import beta
from scipy.integrate import quad
import random

import numpy as np
from matplotlib import pyplot as plt
```

```python
load = 100
```

```python
a, b = 3, 5
loc = 0
scale = 12

beta_dist = beta(a=a, b=b, loc=loc, scale=scale)

# Sample from the distribution and plot the PDF
x = np.linspace(0, 99, 100)
y = beta_dist.pdf(x)
print(x)
print(y, sum(y))
np.random.shuffle(y)
plt.plot(x, y)
plt.show()
```

```python
f = lambda x : beta_dist.pdf(x**0.5)
quad(f, 0, 100)
```
